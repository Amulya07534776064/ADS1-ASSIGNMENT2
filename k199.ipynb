Creating Two Dataframes - Original with Country Names and transposed with Years

```python
import pandas as pd

def read_world_bank_data(filename):
    # Read the data into a dataframe
    df = pd.read_csv(filename, skiprows=4)

    # Clean the dataframes
    df.drop(df.columns[[1,2,3]], axis=1, inplace=True)

    # Transpose the dataframe
    df_transposed = df.transpose()

    return df, df_transposed
```

Creating two frames for the 3 datasets used - Co2 Emission, Forest Land, Arable Land

```python
#co2_df has country names as rows and years as columns
#co2transposed_df has years as rows and country names as columns
co2_df, co2transposed_df = read_world_bank_data("Co2.csv")
forest_df, foresttransposed_df = read_world_bank_data("Forest.csv")
arrable_df, arrabletransposed_df = read_world_bank_data("Arrable.csv")
```

Code to plot Highest/Lowest 5/10 countries for each Factor

```python
import pandas as pd
import matplotlib.pyplot as plt

# Set 'Country' column as the index
co2_df.set_index('Country Name', inplace=True)

# Calculate the row-wise mean for each country
co2_df['Row_Mean'] = co2_df.mean(axis=1)

# Sort the DataFrame based on the mean CO2 emissions and select the top 10
top_10_emitters = co2_df.sort_values(by='Row_Mean', ascending=True).head(5)

# Plot a bar chart for the top 10 highest emitters
plt.figure(figsize=(30,6))
plt.bar(top_10_emitters.index, top_10_emitters['Row_Mean'])
plt.xlabel('Country')
plt.ylabel('Average CO2 Emission')
plt.title('Top 10 Highest CO2 Emitters')
plt.show()
```

Plotting 10 countries Forest Cover/Co2 emission/Arable Land percentages from 1990 to 2019 (5 year increments)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Set 'Country' column as the index
forest_df.set_index('Country Name', inplace=True)

# Select 10 countries for the bar graph
selected_countries = ["China", "Brazil", "India", "France", "Nigeria", "Bangladesh", "South Africa", "Canada", "United Kingdom", "United States"]

# Create a list to store averages for each 5-year interval
averages = []

# Calculate average CO2 emissions for each 5-year interval for selected countries
for start_year in range(1990, 2016, 5):
    end_year = start_year + 4
    selected_years = [str(year) for year in range(start_year, end_year + 1)]
    average_series = forest_df.loc[selected_countries, selected_years].mean(axis=1)
    averages.append(average_series)

# Create a DataFrame from the list of averages
averages_df = pd.concat(averages, axis=1, keys=[f'{start}-{start+4}' for start in range(1990, 2016, 5)])

# Plot a bar graph
averages_df.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Average Forest Cover')
plt.title('Average % Forest Cover for Selected Countries (1990-2020)')
plt.legend(title='5-Year Interval')
plt.show()

```
